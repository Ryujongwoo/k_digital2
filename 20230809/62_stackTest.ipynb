```python
# stackModule 전체를 import 하지 않고 Stack 클래스만 import 해서 사용한다.
from stackModule import Stack
# Stack만 import 시켰으므로 위위 경우처럼 '모듈이름.'을 붙일 필요없이 바로 사용하면 된다.
stack = Stack()
stack.view()
```

```python
menu = 0
while menu != 4:
    while True:
        print('=' * 30)
        print(' 1.push  2.pop  3.view  4.종료 ')
        print('=' * 30)
        menu = int(input('원하는 메뉴를 입력하고 Enter 키를 누르세요: '))
        if menu in [1, 2, 3, 4]:
            break
        else:
            print('메뉴는 1 ~ 4 사이의 정수만 입력해야 합니다.')
        # ===== if
    # ===== while
    
    if menu == 1:
        data = input('스택에 저장할 데이터를 입력하세요: ')
        stack.push(data)
    elif menu == 2:
        stack.pop()
    elif menu == 3:
        stack.view()
    # ===== if
# ===== while
print('프로그램을 종료합니다.')
```
