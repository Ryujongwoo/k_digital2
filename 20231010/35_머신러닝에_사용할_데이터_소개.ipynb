```python
import warnings
warnings.filterwarnings(action='ignore')
%config Completer.use_jedi = False
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 사이킷런 라이브러리에서 제공하는 데이터셋을 사용하기 위해 import 한다.
from sklearn import datasets
```

머신러닝에서 사용할 데이터 소개

집값 예측하기

```python
# 사이킷런 라이브러리에서 보스턴 집값 데이터를 불러온다.
raw_boston = datasets.load_boston()
# print(raw_boston)
```

```python
# 사이킷런에서 제공하는 데이터는 data(피쳐) 부분과 target(레이블, 클래스, 실제값) 부분으로 구성되어 있다.
X_boston = pd.DataFrame(raw_boston.data) # 피쳐 부분을 추출한다.
X_boston.head()
```

```python
Y_boston = pd.DataFrame(raw_boston.target) # 타겟 부분을 추출한다.
Y_boston.head()
```

```python
# 피쳐와 타겟 데이터를 결합해서 전체 데이터프레임을 생성한다.
# 판다스의 concat() 메소드를 이용해서 데이터프레임을 합칠 수 있다.
# concat() 메소드의 axis 속성으로 데이터프레임의 결합 방향을 지정할 수 있고 생략하거나 axis=0으로 설정하면
# 데이터프레임을 위 + 아래 형태로 합치고 axis=1로 설정하면 좌 + 우 형태로 합친다.
df_boston = pd.concat([X_boston, Y_boston], axis=1)
df_boston.head()
```

```python
feature_boston = raw_boston.feature_names
feature_boston
col_boston = np.append(feature_boston, ['target'])
col_boston
```

```python
# 피쳐와 타겟 데이터를 결합한 데이터프레임의 열 이름을 지정한다.
df_boston.columns = col_boston
df_boston.head()
```
